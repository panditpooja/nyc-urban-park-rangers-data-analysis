## Filtering the dataset

##  Datapreprocessing : Data cleaning for the filtered dataset
- Perform data cleaning by changing the datatype of "Date and Time of initial call" from "Object" to "DateTime".
- Column named "Animal Class" contains messy labels with values like (;,#) so in order to keep data consistent and reliable for further analysis, I am replacing the original messy string data by extracting only the label name after # and saving it as new label.


```python
filtered_df['Date and Time of initial call']=pd.to_datetime(filtered_df['Date and Time of initial call'])
```

```python
filtered_df.info()
```

```python
# Split the values in the 'Animal Class' column based on the '#' symbol
filtered_df['Animal Class'] = filtered_df['Animal Class'].str.split('#').str[-1]
```

```python
filtered_df['Report_Year']=filtered_df['Date and Time of initial call'].dt.year
filtered_df
```

```python
filtered_df.drop(columns=['Date and Time of initial call'], inplace=True)
filtered_df
```

## Question 1:
How has the number of animals sent to rehabilitators or animal care centres changed over time?

```python
# Creating a dataframe that holds only information for animals sent to rehabilitators or animal care centres.
rehab_or_acc=final_df[(final_df['Final Ranger Action']=='Rehabilitator') | (final_df['Final Ranger Action']=='ACC')]
rehab_or_acc
```

```python
# Computing the number of animals for each year to understand the annual trend.
no_of_animals_over_time=rehab_or_acc.groupby('Report_Year')['# of Animals'].sum().reset_index()
no_of_animals_over_time
```

```python
# Pivot table for number of animal sent to rehabilitators or animal care centers per year
animal_count_over_time=rehab_or_acc.groupby(['Report_Year','Final Ranger Action'])['# of Animals'].sum().reset_index(name='Animal_Count')
no_of_animals_pivot_table = animal_count_over_time.pivot(index='Final Ranger Action', columns='Report_Year', values='Animal_Count')

# Calculate the total per year
total_per_year = no_of_animals_pivot_table.sum(axis=0)

# Append the total as a new row
no_of_animals_pivot_table.loc['Total'] = total_per_year
no_of_animals_pivot_table

# Define a function to highlight the total row and index
def highlight_total(s):
    is_total = s.name == 'Total'
    return ['background-color: yellow' if is_total else '' for _ in s]

# Define a function to highlight the total index
def highlight_total_index(idx):
    return ['background-color: yellow' if idx == 'Total' else '' for _ in idx]

# Style the DataFrame
styled_table = no_of_animals_pivot_table.style.apply(highlight_total, axis=1)\
                                             .apply(lambda x: ['background-color: yellow' if x.name == 'Total' else '' for _ in x], axis=1)\
                                             .set_table_styles({'Total': [{'selector': 'th', 'props': [('background-color', 'yellow')]}]}, axis=1)

# Display the styled DataFrame
styled_table
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
rehab_or_acc.groupby('Report_Year')['# of Animals'].sum().plot().set_xticks(range(int(rehab_or_acc['Report_Year'].min()), int(rehab_or_acc['Report_Year'].max()) + 1))
plt.title("Total number of animals sent to rehabilitators or animal care centres per year")
plt.xlabel('Year')
plt.ylabel('Number of Animals')
plt.show()
```

## Key Observations:
The lineplot titled "Total number of animals sent to rehabilitators or animal care centres per year" shows the annual trend of animals sent to rehabilitators or care centers from 2018 to 2021.

- **Decrease from 2018 to 2019:**
  - In 2018, the number of animals sent to rehabilitators or animal care centres was at its peak more than 375 animals.
  - In 2019, there was a significant decrease, dropping to  290 animals.

- **Increase from 2019 to 2020:**
  - In 2020, there was a recovery and an increase in the number of animals sent, rising back to 370 animals. This suggests that some factors may have led to more animals being reported or needing care compared to the previous year.
- **Sharp Decline from 2020 to 2021:**
  - The most noticeable change is the sharp decline from 2020 to 2021, where the number dropped to around 200 animals. This is a significant reduction and the lowest number in the observed period.

Although the dataset does not explain what resulted in the significant decline of total number of animals reported annually, as per my subjective opinion/assumptions(no factual proof or data inference), major possibilities could be external factors like environmental changes that influenced wildlife population or the COVID-19 pandemic during 2020 and 2021 likely affected the operations of rehabilitators and care centres.
Strong factual observation about this is not available due to insufficient data pertaining to factors affecting this issue.

**Conclusion:**
The lineplot highlights significant fluctuations, particularly the sharp decline in 2021, warranting further investigation. Understanding these trends is crucial for effective wildlife management and resource allocation.  



## Question 2
How has the type of animal being reported changed over time?

```python
animal_class_over_time=final_df.groupby(['Report_Year','Animal Class']).size().reset_index(name='Animal_Count')
animal_class_over_time
```

```python
# Pivot table for type of animals  being reported per year
animal_class_pivot_table = animal_class_over_time.pivot(index='Animal Class', columns='Report_Year', values='Animal_Count')
animal_class_pivot_table = animal_class_pivot_table.fillna(0).astype(int)

# Calculate the total per year
total_per_year = animal_class_pivot_table.sum(axis=0)  # Summing values in each column (per year)

# Append the total as a new row
animal_class_pivot_table.loc['Total'] = total_per_year
animal_class_pivot_table

# Define a function to highlight the total row and index
def highlight_total(s):
    is_total = s.name == 'Total'
    return ['background-color: yellow' if is_total else '' for _ in s]

# Define a function to highlight the total index
def highlight_total_index(idx):
    return ['background-color: yellow' if idx == 'Total' else '' for _ in idx]

# Style the DataFrame
styled_table = animal_class_pivot_table.style.apply(highlight_total, axis=1)\
                                             .apply(lambda x: ['background-color: yellow' if x.name == 'Total' else '' for _ in x], axis=1)\
                                             .set_table_styles({'Total': [{'selector': 'th', 'props': [('background-color', 'yellow')]}]}, axis=1)

# Display the styled DataFrame
styled_table
```

```python
# Set the style of seaborn
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=animal_class_over_time, x='Report_Year', y='Animal_Count',hue='Animal Class')
plt.title('Count of Animals of each Animal Class by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Animal Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Set the style of seaborn
sns.set(style="whitegrid")

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=animal_class_over_time, x='Report_Year', y='Animal_Count',hue='Animal Class').set_xticks(range(int(animal_class_over_time['Report_Year'].min()), int(animal_class_over_time['Report_Year'].max()) + 1))
plt.title('Count of Animals of each Animal Class by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Animal Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Key Observations:
The bar plot and lineplot both titled "Count of Animals of each Animal Class by Year" shows the annual trend of different types of animals reported from 2018 to 2021.
There are many trends that can be observed out of which the two main ones that I want to highlight are:
- The two most reported animal classes are "Birds" and "Small Mammals-RVS"
- The total number of "Birds" reports increased from 2018 to 2020 and drastically decreased from 2020 to 2021 whereas the total number of "Small Mammals-RVS" continuously decreased each year from 2018 to 2021.

```python
final_df.groupby('Report_Year')['Animal Class'].size().plot().set_xticks(range(int(final_df['Report_Year'].min()), int(final_df['Report_Year'].max()) + 1))
plt.title("Total number of varied animal class reported per year")
plt.xlabel('Year')
plt.ylabel('Number of Animal Class')
plt.show()
```

## Summary statistics for the Animal class over time.
From the above lineplot, it is observed that the total number of animal classes reported over the years is very fluctuating. Highest animal classes were reported in the year 2020 and least in 2021.

## Create a publication table without table analogy
A publication table comprises information that consolidates all the answers for the 3 questions. So concatenating the above 3 pivot tables to get a publication table and storing it as a new sheet in excel file named "Publication Table Without Table Analogy".

## Note:
- This publication table is purely the result of concatenation using python code and will be saved as an excel sheet.
- There is one more sheet in the same excel file named **"Publication Table"** that has all data exactly that of present in excel sheet "Publication Table Without Table Analogy". The only difference is **"Publication Table"** is manually created as a sheet in excel and has proper formatting as per the Table Anatomy guidelines.



```python
publication_table_without_table_analogy= pd.concat([no_of_animals_pivot_table,animal_class_pivot_table,call_source_pivot_table],axis=0).reset_index()
publication_table_without_table_analogy
```

```python
# Save the results to different sheets in an Excel file
output_file = 'Pooja_Output_File.xlsx'
with pd.ExcelWriter(output_file,mode='a') as writer:
    no_of_animals_over_time.to_excel(writer, sheet_name='Number of Animals Over Time', index=False)
    animal_class_over_time.to_excel(writer, sheet_name='Animal Types Over Time', index=False)
    call_source_over_time.to_excel(writer, sheet_name='Call Sources Over Time', index=False)
    publication_table_without_table_analogy.to_excel(writer, sheet_name='Publication Table Without Table Analogy', index=False)

print(f'Results saved to {output_file}')
```
